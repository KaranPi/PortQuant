```python
# --- Make local package importable (robust for notebooks in subfolders) ---
import sys
from pathlib import Path

def add_src_to_path(pkg="quantlib", levels_up=3):
    here = Path.cwd().resolve()
    candidates = [here / "src"]
    # also check parents: ../src, ../../src, ...
    for i in range(1, levels_up+1):
        candidates.append(here.parents[i-1] / "src")
    for p in candidates:
        if (p / pkg / "__init__.py").exists():
            sys.path.insert(0, str(p))
            print(f"[OK] Added to sys.path: {p}")
            return p
    raise FileNotFoundError("Could not find src/quantlib/__init__.py relative to this notebook.")

add_src_to_path()
```

```python
# --- bootstrap & imports ---
import sys, pathlib; sys.path.append(str(pathlib.Path.cwd() / "src"))
from pathlib import Path
import pandas as pd
from quantlib import viz as qviz
DATA_INT = Path(r"C:\Users\quantbase\Desktop\quant\data_int")
FIG_DIR = Path(r"C:\Users\quantbase\Desktop\quant\figures"); FIG_DIR.mkdir(exist_ok=True, parents=True)
```

```python
# --- choose a symbol & gather horizons ---
SYMBOL = "ZOMATO"  # change as needed
r_1d = pd.read_pickle(DATA_INT/"log_returns_daily.pkl")[SYMBOL]
r_21 = pd.read_pickle(DATA_INT/"log_returns_trailing_w21.pkl")[SYMBOL]
r_63 = pd.read_pickle(DATA_INT/"log_returns_trailing_w63.pkl")[SYMBOL]
r_252 = pd.read_pickle(DATA_INT/"log_returns_trailing_w252.pkl")[SYMBOL]

qviz.plot_multi_tf_pdf(
    data={"1D": r_1d.dropna(), "21D": r_21.dropna(), "63D": r_63.dropna(), "252D": r_252.dropna()},
    title=f"{SYMBOL} — PDFs across horizons",
    show_kde=True, show_normal=False, show_student=False,
    save_path=FIG_DIR/f"pdf_{SYMBOL}.png"
)
```
